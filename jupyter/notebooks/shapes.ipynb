```python
import numpy as np

images = np.load("data/shapes/boxes.npy").astype(np.float32)
images = np.append(images, np.load("data/shapes/circles.npy").astype(np.float32), axis=0)
images = np.append(images, np.load("data/shapes/crosses.npy").astype(np.float32), axis=0)
images = np.append(images, np.load("data/shapes/triangles.npy").astype(np.float32), axis=0)
np.random.shuffle(images)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(images[i])
    plt.xlabel(i)
```

```python
import math
import pink

som_dim = 5
neuron_dim = int(images.shape[1] / math.sqrt(2.0) * 2.0)

np_som = np.random.rand(som_dim, som_dim, neuron_dim, neuron_dim).astype(np.float32)
som = pink.som(np_som)

trainer = pink.trainer(som)
```

```python
np_image = images[0].astype(np.float32)
np_image /= 255

image = pink.data(np_image)
trainer(image)
```

```python
from tqdm import tqdm

for i in tqdm(range(1000)):

    np_image = images[i].astype(np.float32)
    np_image /= 255

    image = pink.data(np_image)
    trainer(image)

np_som = np.array(som, copy=True)

plt.figure()
plt.imshow(np_som.swapaxes(1, 2).reshape((som_dim * neuron_dim, som_dim * neuron_dim)))
plt.colorbar()
plt.grid(False)
```
